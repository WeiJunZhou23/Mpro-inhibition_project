```python
import pandas as pd
```

```python
df = pd.read_csv('./data/Asinex_library_inhibition_output.csv')
```

```python
# Inhibition Probability Greater than 0.5
df_positive_prediction = df[df['Binary output (threshold 0.5)'] == 1].copy()
```

```python
df_positive_prediction
```

```python
df_logP_prediction = pd.read_csv('./data/Asinex_KANO_positive_prediction_logP_KANO_prediction.csv')
```

```python
df_logS_prediction = pd.read_csv('./data/Asinex_KANO_positive_prediction_logS_KANO_prediction.csv')
```

```python
logP_prediction = df_logP_prediction['KANO_preds'].tolist()
logS_prediction = df_logS_prediction['KANO_preds'].tolist()
df_positive_prediction['logP'] = logP_prediction
df_positive_prediction['logS'] = logS_prediction
```

```python
df_positive_prediction
```

```python
# 0 < Predicited logP < 5; -5 < Predicted logS < 0; Tanimoto Similarity < 0.85
filtered_output = df_positive_prediction[(df_positive_prediction['logP'] > 0) & (df_positive_prediction['logP'] < 5) & 
                                            (df_positive_prediction['logS'] > -5) & (df_positive_prediction['logS'] < 0) & 
                                                (df_positive_prediction['Tanimoto_sim'] < 0.85)]
```
